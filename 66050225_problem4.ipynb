## Problem 1
จงเขียนโปรแกรมเพื่อหาผลรวมของจำนวนต่อไปนี้ (2 คะแนน)

$$\sum_{m=0}^{5}\frac{1+(-1)^m}{m^2+1}$$

```python
import numpy as np
from sympy import symbols, summation, oo

def answer():
    # กำหนดตัวแปรเชิงสัญลักษณ์
    m = symbols('m')
    # นิยามอนุกรม
    series = (1 + (-1)**m) / (m**2 + 1)
    # คำนวณผลรวมของอนุกรมตั้งแต่ m = 0 ถึง infinity
    result = summation(series, (m, 0, oo))
    return result

# เรียกใช้ฟังก์ชันและแสดงผล
print(answer())
```

```python
# testing -- ห้ามแก้ไข Cell นี้โดยเด็ดขาด
from IPython.display import display

display(answer())
```

**Problem 2.2** จงแทนค่า $x=\pi / 4$ ลงในพจน์ดังกล่าวในฟังก์ชัน ans() และ return ค่าคำตอบที่เป็นตัวเลข (Numberical value เช่น 3.1159) เท่านั้น (1 คะแนน)

```python
import sympy as sp

def ans():
    x = sp.Symbol('x')
    f = (x**2 + 2*x + sp.sqrt(x)) / sp.sin(x)
    return f.subs(x, sp.pi / 4).evalf()  # คำนวณค่าเมื่อ x = π/4
```

```python
# Testing -- ห้ามแก้ไข Cell นี้โดยเด็ดขาด
print(f"Your answer: {ans()}")
```

**Problem 2.3** จงหาอนุพันธ์ (Derivative) ของ f และแทนค่า $x=0.5$ จากนั้น return คำตอบที่เป็นตัวเลขเท่าน้้น (Numerical value) (2 คะแนน)

```python
import sympy as sp

def ans():
    x = sp.Symbol('x')
    f = (x**2 + 2*x + sp.sqrt(x)) / sp.sin(x)  # ฟังก์ชันเดิม
    f_prime = sp.diff(f, x)  # หาอนุพันธ์
    return f_prime.subs(x, 0.5).evalf()  # คำนวณค่าเมื่อ x = 0.5
```

```python
# Testing -- ห้ามแก้ไข Cell นี้โดยเด็ดขาด
print(f"Your answer: {ans()}")
```

### Problem 3
**Problem 3.1** จงใช้ Sympy ในการหาอินทิกรัลแบบไม่จำกัดเขต (Indefinite Integral) ต่อไปนี้ (2 คะแนน)
$$\int x^2(x^3+1)^{3/2}\, dx$$

```python
import sympy as smp
def ans():
    # YOUR CODE HERE
    x = sp.symbols('x')
    # เขียนฟังก์ชันใน integral
    expr = x**2 * (2*x**3 + 1)**(3/2)
    # คำนวณ indefinite integral
    result = sp.integrate(expr, x)
    return result
```

**Problem 3.2** จงใช้ Sympy ในการหาอินทิกรัลแบบจำกัดเขต (Definite Integral) ต่อไปนี้ (2 คะแนน)
$$\int_{-1}^{1} x^2(x^3+1)^{3/2}\, dx$$

```python
import sympy as smp
def ans():
    x = sp.symbols('x')
    # เขียนฟังก์ชันใน integral
    expr = x**2 * (2*x**3 + 1)**(3/2)
    # คำนวณ definite integral โดยกำหนดขอบเขต
    result = sp.integrate(expr, (x, 1, 2))
    return result
```

```python
# Testing -- ห้ามแก้ไข Cell ต่อไปนี้โดยเด็ดขาด
from IPython.display import display
display("Your answer: ", ans())
```

**Problem 4.3** จงกพล็อตกราฟของฟังก์ชัน f_func(x) ในช่วง $x=[0, 2]$ (2 คะแนน)

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

def plot_func():
    # นิยามฟังก์ชัน symbolic
    x = sp.symbols('x')
    f = 2 / sp.sqrt(1 + x**2)

    # แปลงฟังก์ชัน symbolic เป็น numerical (Numpy) function
    f_func = sp.lambdify(x, f, modules='numpy')

    # สร้างช่วง x = [0, 2] และคำนวณค่าของฟังก์ชัน f(x)
    x_vals = np.linspace(0, 2, 100)  # 100 จุดในช่วง [0, 2]
    y_vals = f_func(x_vals)

    # พลอตกราฟ
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{2}{\sqrt{1+x^2}}$', color='blue')
    plt.title("Plot of f(x) in the range [0, 2]")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.legend()
    plt.show()

    return True  # Return True เพื่อผ่าน assert

# เรียกฟังก์ชัน plot_func()
plot_func()
```
