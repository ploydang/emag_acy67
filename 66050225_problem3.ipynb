### Problem 1

สมมติให้มีประจุ $Q_A=-20~\mu\mathrm{C}$ วางอยู่ที่จุด $A(-6, 4, 7)$ และประจุ $Q_B=50~\mu\mathrm{C}$ วางอยู่ที่จุด $B(5, 8, -2)$ ในสุญญากาศ ถ้าระยะที่กำหนดให้มีหน่วยเป็นเมตร จงหา

#### Problem 1.1
(1 คะแนน)

$\mathbf{R_\mathrm{AB}}$

```python

import sympy as smp

from sympy import sqrt # นำเข้า sqrt จาก sympy

def R_AB():
    # พิกัดของจุด A และ B
    x1, y1, z1 = -6, 4, 7 # พิกัดของจุด A
    x2, y2, z2 = 5, 8, -2 # พิกัดของจุด B

    # คำนวณระยะตามสูตร
    delta_x = x2 - x1 # หาค่า x2 - x1
    delta_y = y2 - y1 # หาค่า y2 - y1
    delta_z = z2 - z1 # หาค่า z2 - z1

    # ยกกำลังสองและรวมกัน
    distance_squared = delta_x**2 + delta_y**2 + delta_z**2

    # ถอดรูทหาค่าระยะห่าง
    distance = sqrt(distance_squared)

    return distance # คืนค่าระยะห่าง
```

#### Problem 1.3
(1 คะแนน)

แรงที่กระทำกับประจุ $Q_A$ เนื่องจากประจุ $Q_B$

```python
 from sympy import sqrt, pi, symbols
import sympy as smp

def F_BA():
    # YOUR CODE HERE
      epsilon_0 = 8.85e-12 # ค่าคงที่ไดอิเล็กตริกในสุญญากาศ (F/m)
      Q_A = 20e-6 # ประจุ Q_A (-20e-6)
      Q_B = 50e-6 # ประจุ Q_B (50e-6)

    # พิกัดจุด A และ B
      x1, y1, z1 = -6, 4, 7 # พิกัดของจุด A
      x2, y2, z2 = 5, 8, -2 # พิกัดของจุด B

    # คำนวณ R_AB และเวกเตอร์ R_AB
      delta_x = x2 - x1
      delta_y = y2 - y1
      delta_z = z2 - z1
      R_AB = sqrt(delta_x**2 + delta_y**2 + delta_z**2)

    # เวกเตอร์หน่วย R_hat_AB
      R_hat_AB = (delta_x / R_AB, delta_y / R_AB, delta_z / R_AB)

    # คำนวณแรง F ตามกฏของคูลอมบ์
      F_magnitude = (1 / (4 * pi * epsilon_0)) * (abs(Q_A * Q_B) / R_AB**2)
      F_vector = tuple(F_magnitude * comp for comp in R_hat_AB) # เวกเตอร์แรง

      return F_vector # คืนค่าแรงในรูปเวกเตอร์
```

### Problem 2
ประจุแบบจุดตัวหนึ่งมีขนาด $Q$ วางอยู่ที่ตำแหน่ง $\mathbf{r'}=x'\mathbf{a}_x+y'\mathbf{a}_y+z'\mathbf{a}_z$

#### Problem 2.1
จงสร้างฟังก์ชันที่จะ return ค่าของสนามไฟฟ้าเนื่องจากประจุตัวนี้ ณ ตำแหน่งใด ๆ (1 คะแนน)

```python
import sympy as smp

# Use the following variables
x, y, z = smp.symbols('x y z', real=True)
xp, yp, zp = smp.symbols("x' y' z'", real=True)
Q, epsilon_0, pi = smp.symbols('Q epsilon_0 pi', real=True)

def E():
    # YOUR CODE HERE
     r = smp.Matrix([x, y, z])
     rp = smp.Matrix([xp, yp, zp])

    # Compute the difference between r and rp
     r_diff = r - rp

    # Compute the magnitude of r_diff (||r - r'||)
     r_diff_magnitude = smp.sqrt(r_diff.dot(r_diff))

    # Compute the electric field equation
     E_field = Q / (4 * pi * epsilon_0 * r_diff_magnitude**3) * r_diff

    # Return the electric field vector
     return E_field
```

```python
# Testing - ห้ามแก้ไข Cell นี้โดยเด็ดขาด
from IPython.display import display

display(E())

### BEGIN HIDDEDN TESTS
r = smp.Matrix([x, y, z])
rp = smp.Matrix([xp, yp, zp])

R = r - rp

E_sym = Q / (4 * pi * epsilon_0) * R / R.norm()**3

assert E_sym == E()
```

```python
from scipy.constants import pi, epsilon_0

E_Q = E()

# แก้ไข Function ต่อไปนี้ เพื่อ return คำตอบ
def answer_2_2():
    # YOUR CODE HERE
    E_Q = E()

    # กำหนดค่า Q และตำแหน่งต่าง ๆ
    Q_val = -2.5e-6  # Q = -2.5 μC
    xp_val, yp_val, zp_val = 1, 2, 1  # ตำแหน่งของประจุ
    x_val, y_val, z_val = 0, 0, 0  # จุดที่ต้องการคำนวณสนามไฟฟ้า (ตัวอย่าง)

    # แทนค่าตัวแปรลงในสมการ
    ans = E_Q.subs({
        Q: Q_val,
        epsilon_0: epsilon_0,
        pi: pi,
        x: x_val, y: y_val, z: z_val,
        xp: xp_val, yp: yp_val, zp: zp_val
    })
    return ans
```

### Problem 3

ในระบบพิกัด 2 มิติ มีประจุ 2 ตัว ได้แก่ $Q_1=1~\mu\mathrm{C}$ และ $Q_2=-1~\mu\mathrm{C}$ วางอยู่ที่จุด $(-1, 0)$ และ $(1, 0)$ ตามลำดับ

#### Problem 3.1
จงสร้างฟังก์ชัน E1_func(x, y) และ E2_func(x, y) ซึ่ง return ค่าของเวคเตอร์ ณ ตำแหน่ง $(x, y)$ ใด ๆ ในรูป np.array (2 คะแนน)

```python
import sympy as smp
import numpy as np


# กำหนดตัวแปรที่จะใช้
x, y = smp.symbols('x y', real=True)

# YOUR CODE HERE
k = 9e9  # ค่า 1 / (4 pi epsilon_0)
Q1 = 1e-6  # Q1 = 1 μC
Q2 = -1e-6  # Q2 = -1 μC
x1, y1 = -1, 0  # ตำแหน่ง Q1
x2, y2 = 1, 0   # ตำแหน่ง Q2

# นิยามฟังก์ชันสำหรับ Q1
def E1_func(x, y):
    # เวกเตอร์ r จาก Q1 ไปยังจุด (x, y)
    r_vec = np.array([x - x1, y - y1])
    r_mag = np.linalg.norm(r_vec)  # ความยาว r
    if r_mag == 0:  # ป้องกันหารด้วย 0
        return np.array([0, 0])
    # สนามไฟฟ้าจาก Q1
    E = k * Q1 / r_mag**3 * r_vec
    return E

# นิยามฟังก์ชันสำหรับ Q2
def E2_func(x, y):
    # เวกเตอร์ r จาก Q2 ไปยังจุด (x, y)
    r_vec = np.array([x - x2, y - y2])
    r_mag = np.linalg.norm(r_vec)  # ความยาว r
    if r_mag == 0:  # ป้องกันหารด้วย 0
        return np.array([0, 0])
    # สนามไฟฟ้าจาก Q2
    E = k * Q2 / r_mag**3 * r_vec
    return E
```

#### Problem 3.2
จงสร้างฟังก์ชัน E_func(x, y) ซึ่ง return ค่าของสนามไฟฟ้าเนื่องจากประจุ $Q_1$ และ $Q_2$ ณ ตำแหน่งใด ๆ ในรูปแบบ np.array (3 คะแนน)

```python
def E_func(x, y):
    # YOUR CODE HERE
    k = 8.99e9

    # ค่าประจุและตำแหน่งของ Q1 และ Q2
    Q1, Q2 = 1e-9, -1e-9  # ประจุเป็นคูลอมบ์
    pos_Q1 = np.array([1, 1])   # ตำแหน่ง Q1 (x1, y1)
    pos_Q2 = np.array([-1, -1]) # ตำแหน่ง Q2 (x2, y2)

    # ตำแหน่งที่ต้องการคำนวณ
    P = np.array([x, y])

    # คำนวณสนามไฟฟ้าจาก Q1
    r1 = P - pos_Q1  # เวกเตอร์จาก Q1 ไปยัง P
    r1_mag = np.linalg.norm(r1)  # ระยะทาง
    E1 = k * Q1 / r1_mag**3 * r1  # สนามไฟฟ้า (เวกเตอร์)

    # คำนวณสนามไฟฟ้าจาก Q2
    r2 = P - pos_Q2
    r2_mag = np.linalg.norm(r2)
    E2 = k * Q2 / r2_mag**3 * r2

    # รวมสนามไฟฟ้าทั้งสอง
    E_total = E1 + E2

    # Return เป็น np.array 3 มิติ [Ex, Ey, 0]
    return np.array([E_total[0], E_total[1], 0])

# ทดสอบฟังก์ชัน
print(E_func(0, 0))
```

#### Problem 3.3

จงพล็อตสนามไฟฟ้าเนื่องจากประจุ $Q_1$ และ $Q_2$ โดยใช้ฟังก์ชัน field_plot (5 คะแนน)

**Hint:** ทดลองใช้ option num)_grids = 25

**สำคัญมาก** ให้ return True สำหรับฟังก์ชัน myplot() ด้วย มิฉะนั้น ระบบจะไม่ให้คะแนนนะครับ // Code อาจจะแจ้ง warning ออกมา ในขั้นนี้ไม่เป็นปัญหาครับ

```python
import matplotlib.pyplot as plt

def myplot():
    # YOUR CODE HERE
     k = 8.99e9

    # ประจุและตำแหน่งของ Q1 และ Q2
     Q1, Q2 = 1e-9, -1e-9  # ประจุ Q1 = +1 nC, Q2 = -1 nC
     pos_Q1 = np.array([1, 1])    # ตำแหน่ง Q1 ที่ (1, 1)
     pos_Q2 = np.array([-1, -1])  # ตำแหน่ง Q2 ที่ (-1, -1)

    # สร้างกริดของจุดคำนวณสนามไฟฟ้า
     x = np.linspace(-3, 3, 25)  # แบ่งแกน x เป็น 25 จุด
     y = np.linspace(-3, 3, 25)  # แบ่งแกน y เป็น 25 จุด
     X, Y = np.meshgrid(x, y)

    # ฟังก์ชันคำนวณสนามไฟฟ้า
     def electric_field(Q, pos, X, Y):
        r_x = X - pos[0]
        r_y = Y - pos[1]
        r = np.sqrt(r_x**2 + r_y**2)  # ระยะห่าง
        r_hat_x = r_x / r  # ทิศทาง x
        r_hat_y = r_y / r  # ทิศทาง y
        Ex = k * Q * r_hat_x / r**2  # สนามไฟฟ้าแกน x
        Ey = k * Q * r_hat_y / r**2  # สนามไฟฟ้าแกน y
        return Ex, Ey

    # สนามไฟฟ้าจาก Q1 และ Q2
     Ex1, Ey1 = electric_field(Q1, pos_Q1, X, Y)
     Ex2, Ey2 = electric_field(Q2, pos_Q2, X, Y)

    # รวมสนามไฟฟ้าจากทั้งสองประจุ
     Ex_total = Ex1 + Ex2
     Ey_total = Ey1 + Ey2

    # พล็อตกราฟสนามไฟฟ้า
     plt.figure(figsize=(8, 8))
     plt.quiver(X, Y, Ex_total, Ey_total, color="blue", scale=1e11)
     plt.scatter([pos_Q1[0], pos_Q2[0]], [pos_Q1[1], pos_Q2[1]], color='red', s=100)
     plt.text(pos_Q1[0], pos_Q1[1], 'Q1', color='red', fontsize=12, ha='center')
     plt.text(pos_Q2[0], pos_Q2[1], 'Q2', color='red', fontsize=12, ha='center')
     plt.xlim(-3, 3)
     plt.ylim(-3, 3)
     plt.xlabel("X-axis")
     plt.ylabel("Y-axis")
     plt.title("Electric Field from Q1 and Q2")
     plt.grid()
     plt.show()

     return True

# ทดสอบการพล็อตสนามไฟฟ้า
myplot()
```
