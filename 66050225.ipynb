````python
def main():
    score = input("Enter your score: ")
    score = float(score)
    print(f"Your grade: {get_grade(score)}")

def get_grade(score):
    """
    Returns the grade based on the input score using match-case.
    Parameters:
        score (float): The score (0-100).
    Returns:
        str: Grade (A, B+, B, C+, C, D+, D, or F).
    """
    if not 0.00 <= score <= 100.00:
        return "Invalid score. Please enter a score between 0 and 100."

    match score:
        case s if s >= 80.00:
            return "A"
        case s if s >= 75.00:
            return "B+"
        case s if s >= 70.00:
            return "B"
        case s if s >= 65.00:
            return "C+"
        case s if s >= 60.00:
            return "C"
        case s if s >= 55.00:
            return "D+"
        case s if s >= 50.00:
            return "D"
        case _:
            return "F"

main()
````
